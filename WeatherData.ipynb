##Projection of vector points in 3D space

```python
Berlin = [26.133572, 80.328353, 5.015542];
Frankfurt = [23.161901, 76.886831, 14.021860];
Hamburg = [24.145767, 79.678764, 10.897413];
Nuremberg = [27.114319, 77.825100, 21.273423];
Munich = [24.573419, 74.416920, 21.734649];
Stuttgart = [20.615747, 81.827868, 6.316270];
Hanover = [26.513550, 80.462603, 9.568481];
Saarbruecken = [23.602173, 83.181582, 11.382041];
Cologne = [27.769321, 79.759665, 11.433353];
Constance = [29.343985, 78.814028, 17.818053];
Freiburg = [25.554123, 77.339895, 7.965502];
Karlsruhe = [19.780618, 76.517790, 6.304491];

Cities = [Berlin, Frankfurt, Hamburg, Nuremberg, Munich, Stuttgart, Hanover, Saarbruecken, Cologne, Constance, Freiburg, Karlsruhe];
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.set_xlabel('Temperature')
ax.set_ylabel('Humidity')
ax.set_zlabel('Wind')

ax.set_xlim(20, 30)
ax.set_ylim(60, 85)
ax.set_zlim(0, 25)

#Scatterplot of all vectors in 3D space
for city in Cities:
    ax.scatter3D(city[0], city[1], city[2], color='b');

# - Line Vector -
LineVector = [40, 120, 0];
ax.quiver(0, 0, 0, LineVector[0], LineVector[1], LineVector[2]);

#Projection onto a line
v = (pow(LineVector[0], 2) + pow(LineVector[1], 2) + pow(LineVector[2], 2)); #magnitude of LineVector using distance formula (squared)

#Using the projection on a line equation... proj a = (a*v / ||v||^2)v
def projection(vector):
    projvector = [None, None, None];
    AV = ((vector[0] * LineVector[0]) + (vector[1] * LineVector[1]) + (vector[2] * LineVector[2]));
    temp = AV/v;
    projvector = ([LineVector[0] * temp, LineVector[1] * temp, LineVector[2] * temp]);
    return projvector;

#Projected vectors onto LineVector <40, 120, 0>
for city in Cities:
    ax.scatter3D(projection(city)[0], projection(city)[1], projection(city)[2], color='r');

#for city in Cities:
    city = Karlsruhe;
    x = [city[0], projection(city)[0]];
    y = [city[1], projection(city)[1]];
    z = [city[2], projection(city)[2]];
    ax.plot(x, y, z, color='g');
    projcity = [projection(city)[0], projection(city)[1], projection(city)[2]];

#Projected vectors
# Berlin = [26.7118631, 80.1355893, 0]; (correct)
# Frankfurt = [25.3822394, 76.1467182, 0]; (correct)
# Hamburg = [26.3182059, 78.9546177, 0]; (correct)
# Nuremberg = [26.0589619, 78.1768857, 0]; (correct)
# Munich = [24.7824179, 74.3472537, 0]; (correct)
# Stuttgart = [26.6099351, 79.8298053, 0]; (correct)
# Hanover = [26.7901359, 80.3704077, 0]; (correct)
# Saarbruecken = [27.3146919, 81.9440757, 0]; (correct)
# Cologne = [26.7048316, 80.1144948, 0]; (correct)
# Constance = [26.5786069, 79.7358207, 0]; (correct)
# Freiburg = [25.7573808, 77.2721424, 0]; (correct)
# Karlsruhe = [24.9333988, 74.8001964, 0]; (correct)

#ax.view_init(10,10,10);

#Key:
# Blue -> Plotted vector points (cities)
# Red -> Projected vector points (cities) on a line vector <40, 120, 0>
```
